```python
def rounder(num):  
    if (float(num) % 1) >= 0.5:
        x = np.ceil(num)
    else:
        x = np.floor(num)
    return x

count=0
sumS=0
for i in series:
    number=rounder(i)
    series[count]=number
    count+=1
    sumS+=number
print(series/5)
print("Sum:", sumS)
```
